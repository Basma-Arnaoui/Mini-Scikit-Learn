```python
import sys
sys.path.append("../")
import numpy as np
from sklearn.model_selection import KFold as SklearnKFold
from model_selection.KFold import KFold

def test_k_fold():
    data = np.random.rand(100, 2)  # Generate random data for more robust testing

    # Custom KFold
    custom_kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    custom_indices = list(custom_kfold.split(data))

    # Sklearn KFold
    sklearn_kfold = SklearnKFold(n_splits=5, shuffle=True, random_state=42)
    sklearn_indices = list(sklearn_kfold.split(data))

    print(custom_indices)
    print("============")
    print(sklearn_indices)

    print("Custom KFold tests passed successfully!")

test_k_fold()
```
